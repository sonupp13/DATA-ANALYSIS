**MERGING**

```python
import pandas as pd
data1 = pd.DataFrame(
  {
      'key1': [1,2,4,5,6],
      'key2': [4,5,6,7,8],
      'key3': [3,4,5,6,6]
}
)
data2 = pd.DataFrame(
        {'key1': [1,2,45,6,67],
        'key4': [56,5,6,7,8],
        'key5': [3,56,5,6,6]
}
)

s= pd.merge(data1,data2)
s
```

```python
pd.merge(data1,data2 , how='left')
```

```python
pd.merge(data1,data2 , how='right')
```

```python
pd.merge(data1,data2 , how='cross')
```

```python
import pandas as pd

 data1= pd.DataFrame({'key1': [1,2,4,5,6],
 'key2': [4,5,6,7,8],
 'key3': [3,4,5,6,6]},
 index = ['a','b','c','d','e']
)

data2= pd.DataFrame({'key11': [1,2,45,6,67],
 'key22' :[6,5,6,7,8],
 'key33':[56,5,6,6,8]},
 index = ['a','b','h','j','k']
)

data1.join(data2)
```
