```python
data1 = pd.DataFrame(
  {
      'key1': [1,2,4,5,6],
      'key2': [4,5,6,7,8],
      'key3': [3,4,5,6,6]
}
)
```

```python
data1
```

```python
data2 = pd.DataFrame(
        {'key1': [1,2,45,6,67],
        'key4': [56,5,6,7,8],
        'key5': [3,56,5,6,6]
}
)
```

```python
data2
```

```python
#merge() function in pandas is used to merge two DataFrames based on a
#common column or index
pd.merge(data1,data2)
```

```python
pd.merge(data1,data2 , how='left')    #Left Join

#All Rows from Left: All rows from the left DataFrame (data1) are retained.
#Matching Rows from Right: Rows from the right DataFrame (data2) that match with the left DataFrame are included.
#No Match Handling: If there is no match, the columns from the right DataFrame will contain NaN (missing) values in the resulting DataFrame (merged_data_left).
```

```python
pd.merge(data1, data2, how='right')
```

```python
pd.merge(data1,data2 , how='cross')  #cross product
```

```python
data1
```

```python
data2
```

```python
pd.concat([data1,data2],axis=1)
```

```python
data1 = pd.DataFrame({'key1': [1,2,4,5,6],
 'key2': [4,5,6,7,8],
 'key3': [3,4,5,6,6]},
 index = ['a','b','c','d','e']
)
```

```python
data2 = pd.DataFrame({'key11': [1,2,45,6,67],
 'key22' :[6,5,6,7,8],
 'key33':[56,5,6,6,8]},
 index = ['a','b','h','j','k']
)
data2
```

```python
data1
```

```python
data1.join(data2)
```

```python
data1.join(data2).fillna('abhi')
```

```python
data1.join(data2 , how='right')
```

```python
 #inner join
 #Only the rows with matching keys in both DataFrames (data1 and data2) are included.
data1.join(data2 , how='inner')
```

```python
data1.join(data2 , how='outer')
#Outer Join:
#All Rows from Both: All rows from both DataFrames (data1 and data2) are included.
#Matching Rows: Rows with common keys are included.
#NaN Values for Non-Matches: If there is no match, the columns from the non-matching DataFrame will contain NaN values.
```
